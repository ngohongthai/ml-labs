## Làm việc với file PDF
#### 4. Sử dụng PyPDF2 để mở file `Business_Proposal.pdf`. Trích xuất text của trang 2.

```python
# Thực hiện import
import PyPDF2

# Mở file ở dạng đối tượng nhị phân
f = open('Business_Proposal.pdf', 'rb')

# Sử dụng PyPDF2 để đọc text của file
pdf_reader = PyPDF2.PdfFileReader(f)


# Lấy text từ trang 2 (THỬ THÁCH: thực hiện điều này trong một bước!)
page_two_text = pdf_reader.getPage(1).extractText()



# Đóng file
f.close()

# In ra nội dung của page_two_text
print(page_two_text)
```

#### 5. Mở file `contacts.txt` ở append mode. Thêm text của trang 2 ở trên vào `contacts.txt`.

#### THỬ THÁCH: Hãy xem liệu bạn có thể loại bỏ từ "AUTHORS:" không.

```python
%%writefile contacts.txt
First_Name Last_Name, Title, Extension, Email
```

```python
# Giải pháp thông thường:
with open('contacts.txt', 'a+') as f:
    f.write(page_two_text)
    
with open('contacts.txt', 'r') as f:
    print(f.read())
```

```python
%%writefile contacts.txt
First_Name Last_Name, Title, Extension, Email
```

```python
# Giải pháp nâng cao (chạy lại cell %%writefile ở trên để lấy file contacts.txt không bị chỉnh sửa):
page2_lines = page_two_text.split('\n')
with open('contacts.txt', 'a+') as f:
    for line in page2_lines[1:]:
        f.write(line + '\n')

with open('contacts.txt', 'r') as f:
    print(f.read())
```

## Regular expressions
#### 6. Sử dụng biến `page_two_text` đã tạo ở trên, trích xuất bất kỳ địa chỉ email nào có trong file `Business_Proposal.pdf`.

```python
import re

# Nhập regex pattern ở đây. Điều này sẽ cần vài lần thử!
pattern = re.compile(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b')

re.findall(pattern, page_two_text)
```
