**Dưới đây là một số code hữu ích:**

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
def display_img(img,cmap=None):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap)
```

**TASK: Mở và hiển thị ảnh giaraffes.jpg nằm trong folder DATA.** 

```python
img = cv2.imread('../DATA/giraffes.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
display_img(img)
```

**TASK: Áp dụng ngưỡng nhị phân vào ảnh.** 

```python
img = cv2.imread('../DATA/giraffes.jpg', 0)
ret, thread = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
display_img(thread, cmap='gray')
```

**TASK: Mở file giaraffes.jpg từ folder DATA, chuyển không gian màu của nó thành HSV và hiển thị ảnh.** 

```python
img = cv2.imread('../DATA/giraffes.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
display_img(img)
```

**TASK: Tạo low pass filter với 4x4 kernel chứa các giá trị 1/10 (0.01) rồi sử dụng 2-D Convolution để làm mờ ảnh hươu (hiển thị ở RGB chuẩn).** 

```python
kenel = np.ones((4,4),np.float32)/10
```

```python
kenel
```

```python
img = cv2.imread('../DATA/giraffes.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
dst = cv2.filter2D(img,-1,kenel)
display_img(dst)
```
