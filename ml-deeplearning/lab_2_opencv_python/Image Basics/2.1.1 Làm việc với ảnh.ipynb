#### TASK: Mở ảnh *dog_backpack.jpg* từ folder DATA và hiển thị trong notebook. Đảm bảo sửa theo thứ tự RGB.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import warnings
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
image = cv2.imread('../DATA/dog_backpack.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

#### TASK: Lật ngược ảnh và hiển thị trong notebook.

```python
flipped = cv2.flip(image, 0)
plt.imshow(flipped)
```

#### TASK: Vẽ một hình chữ nhật trống MÀU ĐỎ quanh khuôn mặt của chú chó và hiển thị trong notebook.

```python
image_reg = cv2.rectangle(image.copy(), pt1=(200, 380), pt2=(600, 700), color=(255, 0, 0), thickness=10)
plt.imshow(image_reg)
```

#### TASK: Vẽ một HÌNH TAM GIÁC MÀU XANH DƯƠNG ở giữa ảnh. Có thể chọn kích thước và góc tùy ý, nhưng phải là hình tam giác (3 cạnh) theo bất kỳ hướng nào.

```python
vertices = np.array([[250, 700], [425, 400], [600, 700]], np.int32)
image_tri = cv2.polylines(image.copy(), [vertices], isClosed=True, color=(0, 0, 255), thickness=10)
plt.imshow(image_tri)
```

### BONUS TASK. Bạn có thể tô màu tam giác này không? Điều này cần một hàm khác, không có trong bài giảng. Hãy thử xem liệu bạn có thể tìm thấy trên Google không.

[CLICK VÀO LIÊN KẾT SAU ĐỂ XEM GỢI Ý](https://docs.opencv.org/3.0-beta/modules/imgproc/doc/drawing_functions.html#fillpoly)

```python
image_fillpoli = cv2.fillPoly(image.copy(), [vertices], color=(0, 0, 255))
plt.imshow(image_fillpoli)
```
